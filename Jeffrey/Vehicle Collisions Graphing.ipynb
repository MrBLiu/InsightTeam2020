```python
#data handling/modeling
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
import scipy.stats as stats

# visualization
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Motor_Vehicle_Crashes.csv")
print(data.dtypes)
print(data.shape)
```

```python
feature_cols = data.columns[10:18]
```

```python
x = data[feature_cols].copy()
x
```

```python
x = x[x['NUMBER OF PERSONS INJURED']>0]
x = x.dropna()
pred = x['NUMBER OF PERSONS INJURED']
resp = x['NUMBER OF PERSONS KILLED']
pred.dropna()
resp.dropna()
pred = np.array(x['NUMBER OF PERSONS INJURED'])
resp = np.array(x['NUMBER OF PERSONS KILLED'])
pred = pred.reshape(-1,1)
resp = resp.reshape(-1,1)
#print(x['NUMBER OF PERSONS INJURED'].isnull() = True)
```

```python
linreg = LinearRegression()
linreg.fit(pred,resp)
resp_pred = linreg.predict(pred)
print("The slope is:",linreg.coef_,"The y-intercept is",linreg.intercept_)
print("The R^2 is:",metrics.r2_score(resp,resp_pred))
print("The MSE is:",metrics.mean_squared_error(resp,resp_pred))
```

```python
new_data = pd.DataFrame()
cols_of_interest = data[['CRASH TIME','BOROUGH','NUMBER OF PERSONS INJURED','NUMBER OF PERSONS KILLED','CONTRIBUTING FACTOR VEHICLE 1','CONTRIBUTING FACTOR VEHICLE 2']]
new_data = cols_of_interest
#is it right to drop NA values this early? What is being dropped
new_data.dropna()
```

```python
def borough_to_numeric(x):
    if x=='MANHATTAN':
        return 1
    if x=='BROOKLYN':
        return 2
    if x=='QUEENS':
        return 3
    if x=='BRONX':
        return 4
    if x=='STATEN ISLAND':
        return 5
    else:
        return 0
#creates dummy variable 1 for day and 0 for night  
def time_to_day_night(x):
    h,m = x.split(':')
    h = int(h)
    m = int(m)
    if h >= 12:
        return 1
    else:
        return 0
    
```

```python
new_data['CRASH TIME'] = new_data['CRASH TIME'].apply(time_to_day_night)
new_data['BOROUGH'] = new_data['BOROUGH'].apply(borough_to_numeric)
new_data 
```

```python
new_data['NUMBER OF PERSONS AFFECTED'] = new_data['NUMBER OF PERSONS KILLED'] + new_data['NUMBER OF PERSONS INJURED']
new_data.tail(50)
```

```python
feature_list = ['CRASH TIME','BOROUGH']
x_mult = new_data[feature_list]
y_mult = new_data['NUMBER OF PERSONS AFFECTED']

mult_linreg = LinearRegression()
mult_linreg.fit(x_mult,y_mult)

print('The y-intercept is:', mult_linreg.intercept_)
```
