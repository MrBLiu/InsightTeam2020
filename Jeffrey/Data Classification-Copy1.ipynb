```python
data = pd.read_csv("Motor_Vehicle_Crashes.csv")
def borough(x):
    if x=="QUEENS":
        return 1
    if x=='BROOKLYN':
        return 2
    if x=='MANHATTAN':
        return 3
    if x=='BRONX':
        return 4
    if x=='STATEN ISLAND':
        return 5
data['BOROUGH']=data['BOROUGH'].apply(borough)
print(data.dtypes)
print(data.shape)
```

```python
pd.options.display.max_rows = 150
value_counts = data["VEHICLE TYPE CODE 1"].value_counts(ascending=True)
remove = value_counts[value_counts <= 20].index
data = data[~data['VEHICLE TYPE CODE 1'].isin(remove)]
print(print(data['VEHICLE TYPE CODE 1'].value_counts()))
```

```python
data['VEHICLE TYPE CODE 1'] = data['VEHICLE TYPE CODE 1'].replace(['PASSENGER VEHICLE','Sedan','SPORT UTILITY / STATION WAGON','Station Wagon/Sport Utility Vehicle','Convertible','4 dr sedan','2 dr sedan'],'Car')
data['VEHICLE TYPE CODE 1'] = data['VEHICLE TYPE CODE 1'].replace(['TAXI','Taxi'],'Taxi')
data['VEHICLE TYPE CODE 1'] = data['VEHICLE TYPE CODE 1'].replace(['PICK-UP TRUCK','Pick-up Truck','Box Truck'],'Truck')
data['VEHICLE TYPE CODE 1'] = data['VEHICLE TYPE CODE 1'].replace(['Bus','Tractor Truck Diesel','Garbage or Refuse','Tractor Truck Gasoline','Tow Truck / Wrecker','Tanker','FIRE TRUCK','Concrete Mixer','Beverage Truck','School Bus',],'Large Vehicle')
print(print(data['VEHICLE TYPE CODE 1'].value_counts()))
```

```python
value_counts = data["CONTRIBUTING FACTOR VEHICLE 1"].value_counts(ascending=True)
remove = value_counts[value_counts <= 20].index
data = data[~data['CONTRIBUTING FACTOR VEHICLE 1'].isin(remove)]
#print(print(data['CONTRIBUTING FACTOR VEHICLE 1'].value_counts()))

```
